```python
from scipy.stats import invgauss,norm
from math import sqrt
 

def generate_next_sample(current_sample = None, tau=.01,sigma=1):
    
        
    
    
    if current_sample is None:
        return 0, 0
    else:
        T = invgauss(mu=tau)
        dt = T.rvs()
        X = norm(loc=0, scale=sigma)
        t, xt = current_sample
        t += dt
        xt += X.rvs() * sqrt(dt)
        return t, xt  
        
    
    
    
```

```python
t,xt = generate_next_sample()
```

```python
t,xt = generate_next_sample(current_sample=(t,xt))
```

```python
tt = 0
xt = 0

while True:
    (tt,xt) = generate_next_sample(current_sample=(tt,xt))
    if tt<10:
        print (tt,xt)
    
    else:
        break    
```
